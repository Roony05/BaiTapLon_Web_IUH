```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
# Cấu hình Chrome chạy ở chế độ headless
chrome_options = Options()
chrome_options.add_argument("--headless=new")  # Chế độ không giao diện
chrome_options.add_argument("--no-sandbox")  # Chạy trong môi trường Colab
chrome_options.add_argument("--disable-dev-shm-usage")  # Hạn chế lỗi bộ nhớ
driver = webdriver.Chrome()

```

```python
def crawl_link(driver):
    # Lấy danh sách các phần tử sản phẩm (giả định mỗi sản phẩm là 1 <div> trong .product_grid)
    products = driver.find_elements(By.CSS_SELECTOR, ".product_grid > div")

    # Lấy 4 sản phẩm đầu tiên (hoặc ít hơn nếu không đủ)
    for product in products[:4]:
        try:
            # Tìm thẻ <a> bên trong mỗi sản phẩm
            a_tag = product.find_element(By.TAG_NAME, 'a')
            href = a_tag.get_attribute('href')
            # Tìm thẻ hình ảnh và lấy link ảnh ra 
            img_tag = product.find_element(By.TAG_NAME, 'img')
            img_src = img_tag.get_attribute('src')

            crawl_product(href,img_src)  # Mở liên kết trong trình duyệt
            
            driver.back()  # Quay lại trang trước đó
            print(href)
        except:
            print("Không tìm thấy thẻ <a> trong sản phẩm.")
```

```python
def crawl_product(link_sp,img_src):
    driver.get(link_sp)
    time.sleep(2)  # Đợi trang tải xong
    # L
    name = driver.find_element(By.XPATH, '/html/body/div[10]/div[1]/div[1]/div[2]/div[1]/div[1]/div/h1').text
    print(name)
    price = driver.find_element(By.XPATH, '//*[@id="price"]').text
    print(price)
    # Lấy thông tin sản phẩm
    thuong_hieu = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[1]/td[1]/span/a').text
    print(thuong_hieu)
    xuat_xu = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[1]/td[2]/span').text
    print(xuat_xu)
    doi_tuong = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[2]/td[1]/span/a').text
    print(doi_tuong)
    dong_san_pham = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[2]/td[2]/span').text
    print(dong_san_pham)
    khang_nuoc = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[3]/td[1]/span').text
    print(khang_nuoc)
    loai_may = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[3]/td[2]/span').text
    print(loai_may)
    chat_lieu_kinh = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[4]/td[1]/span').text
    print(chat_lieu_kinh)
    chat_lieu_day = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[4]/td[2]/span').text
    print(chat_lieu_day)
    size_mat = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[5]/td[1]/span').text
    print(size_mat)
    do_day = driver.find_element(By.XPATH, '//*[@id="tab_content_description"]/div/div[2]/div/table/tbody/tr[5]/td[2]/span').text
    print(do_day)

    data_product = {
        'name': name,
        'price': price,
        'img_src': img_src,
        'thuong_hieu': thuong_hieu,
        'xuat_xu': xuat_xu,     
        'doi_tuong': doi_tuong,
        'dong_san_pham': dong_san_pham,
        'khang_nuoc': khang_nuoc,
        'loai_may': loai_may,
        'chat_lieu_kinh': chat_lieu_kinh,
        'chat_lieu_day': chat_lieu_day,
        'size_mat': size_mat,
        'do_day': do_day
    }
    print(data_product)
    return data_product
```

```python
# Tự động tải và dùng driver phù hợp
links = ['https://www.watchstore.vn/dong-ho-nam-chinh-hang-c1','https://www.watchstore.vn/dong-ho-nu-chinh-hang-c1','https://www.watchstore.vn/dong-ho-luxury-chinh-hang-c1','https://www.watchstore.vn/dong-ho-luot-cao-cap-c1']
for link in links:
    # Mở trang web cần cào dữ liệu
    driver.get(link)
    time.sleep(3)
    crawl_link(driver)
```
